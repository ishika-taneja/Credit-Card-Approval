```python
import pandas as pd
```

```python
#loading the dataset
cc_apps = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/credit-screening/crx.data",header=None)
```

```python
cc_apps.head(5)
```

```python
cc_apps_description = cc_apps.describe()
print(cc_apps_description)
```

```python
cc_apps_info = cc_apps.info()
print(cc_apps_info)
```

```python
cc_apps.tail(17)
#we observe a missing value as ?
```

```python
import numpy as np
cc_apps.replace("?",np.NaN,inplace=True) #replacing missing values represented by '?' to 'NaN'
```

```python
cc_apps.tail(17)
#we see that ? has been replaced by NaN
```

```python
cc_apps.fillna(cc_apps.mean(numeric_only=True),inplace=True)
```

```python
cc_apps.isnull().sum()
```

```python
for col in cc_apps.columns:
    if cc_apps[col].dtypes == 'object':
        cc_apps = cc_apps.fillna(cc_apps[col].value_counts().index[0])
```

```python
cc_apps.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for col in cc_apps.columns.values:
    if cc_apps[col].dtypes == 'object':
        cc_apps[col]=le.fit_transform(cc_apps[col])
```

```python
cc_apps.drop([11,13],axis=1,inplace=True)
```

```python
cc_apps = cc_apps.values #converting dataframe to numpy array
```

```python
type(cc_apps)
```

```python
X,y = cc_apps[:,0:13],cc_apps[:,13]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state = 42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))
rescaledX_train = scaler.fit_transform(X_train)
rescaledX_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(rescaledX_train,y_train)
```

```python
y_pred = logreg.predict(rescaledX_test)
```

```python
print("Accuracy of logistic regression classifier: ", logreg.score(rescaledX_test,y_test))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred))
```
